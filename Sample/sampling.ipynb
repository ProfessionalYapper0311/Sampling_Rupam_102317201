```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import math
```

```python
cred = pd.read_csv("Creditcard_data.csv")
cred.sample(5)
```

```python
cred.tail()
```

```python
# for accuracy split original data into Train and test set, before this I attempted it by oversampling the minority class before splitting, which leads to data leakage
X_orig = cred.drop('Class', axis=1)
y_orig = cred['Class']
X_train_orig, X_test, y_train_orig, y_test = train_test_split(X_orig, y_orig, test_size=0.25, random_state=42, stratify=y_orig)
train_df = pd.concat([X_train_orig, y_train_orig], axis=1)
```

```python
maj_class = train_df[train_df['Class'] == 0]
min_class = train_df[train_df['Class'] == 1]
maj_class
```

```python
min_class
```

```python
min_oversample = min_class.sample(len(maj_class), replace=True, random_state=42)
bal_data = pd.concat([maj_class, min_oversample])
bal_data = bal_data.sample(frac=1, random_state=42).reset_index(drop=True)

print(f'Balanced Training Data shape: {bal_data.shape}')
print(f"Class distribution:\n {bal_data['Class'].value_counts()}")

```

```python
Z = 1.96
p = 0.5
e = 0.05
n = math.ceil((Z**2 * p * (1-p)) / e**2)
print(f"Sample size: {n}")

samples = {}

#random
samples['Rndm_Sampling'] = bal_data.sample(n, random_state=42)

#sydtematic
k = int(len(bal_data)/n)
indices = np.arange(0, len(bal_data), k)[:n]
samples['Syst_Sampling'] = bal_data.iloc[indices]

#stratify
sample3, _ = train_test_split(bal_data, train_size=n, stratify=bal_data['Class'], random_state=42)
samples['Stat_Sampling3'] = sample3

#cluster
bal_data_cluster = bal_data.copy()

num_cluster = 20
bal_data_cluster['Cluster'] = np.random.randint(0, num_cluster, size=len(bal_data))
sel_cluster = np.random.choice(num_cluster, size=5, replace=False)
cluster_samp = bal_data_cluster[bal_data_cluster['Cluster'].isin(sel_cluster)]

if len(cluster_samp) > n:
    cluster_samp = cluster_samp.sample(n, random_state=42)
samples['Cluster_Sampling'] = cluster_samp.drop(columns=['Cluster'])

#boorstrap
samples['Btstrap_sampling'] = bal_data.sample(n, replace=True, random_state=42)

print("Samples created")
for name, df in samples.items():
    print(f"{name} Shape: {df.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

models = {
    'M1(LogisticRegression)': LogisticRegression(max_iter=1000, random_state=42),
    'M2(DecisionTreeClassifier)': DecisionTreeClassifier(random_state=42),
    'M3(RandomForestClassifier)': RandomForestClassifier(random_state=42),
    'M4(SVM)': SVC(random_state=42),
    'M5(KNN)': KNeighborsClassifier()
}

print(f"Test Set Size: {len(X_test)} rows")

final_res = {}

for model_name, model in models.items():
    print(f"Training {model_name}...")
    model_accuracy = {}
    
    for sample_name, sample_df in samples.items():
        #specific sample
        X_train_sample = sample_df.drop('Class', axis=1)
        y_train_sample = sample_df['Class']
        
        model.fit(X_train_sample, y_train_sample)
        
        #org untouched x_test
        y_pred = model.predict(X_test)
        
        acc = accuracy_score(y_test, y_pred)
        model_accuracy[sample_name] = round(acc * 100, 2)
    
    final_res[model_name] = model_accuracy

print("\nAccuracy Table:")
results_df = pd.DataFrame(final_res).T
print(results_df)
```

```python
results_df.to_csv("sampling_rupam_102317201.csv")

print("\nBest Sampling Technique for each Model:")
for model in final_res:
    best_sample = max(final_res[model], key=final_res[model].get)
    print(f"{model}: {best_sample} ({final_res[model][best_sample]}%)")
```

```python
best_model_overall = None
best_sample_overall = None
max_accuracy = 0.0

for model_name, accuracies in final_res.items():
    for sample_name, acc in accuracies.items():
        if acc > max_accuracy:
            max_accuracy = acc
            best_model_overall = model_name
            best_sample_overall = sample_name

print("\n" + "="*40)
print("Overall BEST Model and sampling")
print("="*40)
print("\n")
print(f"Best Model         : {best_model_overall}")
print(f"Best Sampling method : {best_sample_overall}")
print(f"Highest Accuracy   : {max_accuracy}%")
```
